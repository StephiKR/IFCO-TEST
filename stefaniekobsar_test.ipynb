```python
def extract_client_name(contact_data):
    if pd.isna(contact_data):  # Verify if it's a NAN value
        return "John Doe"       # If Nan, return "John Doe"
    try:
        contact_list = json.loads(contact_data)
        contact = contact_list[0]
        return f"{contact['contact_name']} {contact['contact_surname']}"
    except (json.JSONDecodeError, KeyError, IndexError) as e:
        return "Invalid Data"  # En caso de error en el formato JSON

# Aplicar la funciÃ³n
orders_copy = orders.copy()
orders_name_client = orders_copy[['order_id', 'contact_data']]
orders_name_client['client_name'] = orders_name_client['contact_data'].apply(extract_client_name)

# Mostrar el DataFrame final
orders_name_client[['order_id', 'client_name']]
```

### Test 3: DataFrame of Orders with Contact Address

Provide a DataFrame (`df_2`) containing the following columns:

```python
def extract_contact_address(contact_data):
    if pd.isna(contact_data):  # Verifica si es NaN
        return "Unknown, UNK00"
    try:
        contact_list = json.loads(contact_data)
        contact = contact_list[0]
        
        # Obtener ciudad y cÃ³digo postal, aplicando placeholders si falta informaciÃ³n
        city = contact.get('city') if contact.get('city') else "Unknown"
        cp = contact.get('cp') if contact.get('cp') else "UNK00"
        
        return f"{city}, {cp}"
    except (json.JSONDecodeError, KeyError, IndexError):
        return "Unknown, UNK00"

# Aplicar la funciÃ³n para crear 'contact_address'
df_2 = orders.copy()
df_2['contact_address'] = orders_copy['contact_data'].apply(extract_contact_address)

# Crear el DataFrame final con las columnas requeridas
df_2 = df_2[['order_id', 'contact_address']]

df_2
```

### Test 4: Calculation of Sales Team Commissions

The Sales Team requires your assistance in computing the commissions. It is possible for multiple salespersons to be associated with a single order, as they may have participated in different stages of the order. The `salesowners` field comprises a ranked list of the salespeople who have ownership of the order. The first individual on the list represents the primary owner, while the subsequent individuals, if any, are considered co-owners who have contributed to the acquisition process. The calculation of commissions follows a specific procedure:

- Main Owner: 6% of the net invoiced value.
- Co-owner 1 (second in the list): 2.5% of the net invoiced value.
- Co-owner 2 (third in the list): 0.95% of the net invoiced value.
- The rest of the co-owners do not receive anything.

Provide a list of the distinct sales owners and their respective commission earnings. The list should be sorted in order of descending performance, with the sales owners who have generated the highest commissions appearing first.

**Hint:** Raw amounts are represented in cents. Please provide euro amounts with two decimal places in the results.

**Include unit tests** to verify the correctness of the commission calculations and sorting order.


```python
with open('resources/invoicing_data.json', 'r') as f: #open de json as a dictionary
    json_data = json.load(f)

# Accede a la clave 'data' y luego a 'invoices'
invoices_data = json_data['data']['invoices']

# Convierte la lista de facturas en un DataFrame
invoices_df = pd.DataFrame(invoices_data)

# Muestra las primeras filas
print(invoices_df.shape)
invoices_df.head()
```

```python
invoices_df = invoices_df.rename(columns={'orderId': 'order_id'}) #rename column orderId to order_id to match with orders_copy
merged_raw_df = pd.merge(invoices_df, orders_copy, on='order_id', how='inner') #inner join to merge both dataframes
merged_raw_df[['Main_Owner', 'Co_owner_1', 'Co_owner_2']] = merged_raw_df['salesowners'].str.split(', ', expand=True, n=2) #split salesowners column into 3 columns 
merged_raw_df.drop(columns=['salesowners'], inplace=True) #drop salesowners column  

merged_raw_df.head()
```

```python
sales_owners_value = merged_raw_df.copy() 
sales_owners_value.head()
```

```python
sales_owners_value['grossValue'] = sales_owners_value['grossValue'].astype(int)
sales_owners_value['vat'] = sales_owners_value['vat'].astype(int)

sales_owners_value.info()
```

```python
sales_owners_value['net_value'] = sales_owners_value['grossValue'] * (1 - (sales_owners_value['vat']/100))
sales_owners_value.head(15)
```

```python
commissions_df = sales_owners_value[['Main_Owner', 'Co_owner_1', 'Co_owner_2', 'net_value']].copy()
```

```python
commissions = {}

for _, row in commissions_df.iterrows():
    net_value = row["net_value"] / 100.0  # Convertir a euros
    
    main_owner = row["Main_Owner"]
    co_owner_1 = row["Co_owner_1"]
    co_owner_2 = row["Co_owner_2"]
    
    commissions[main_owner] = round(commissions.get(main_owner, 0) + net_value * 0.06, 2)
    
    if pd.notna(co_owner_1):
        commissions[co_owner_1] = round(commissions.get(co_owner_1, 0) + net_value * 0.025, 2)
    
    if pd.notna(co_owner_2):
        commissions[co_owner_2] = round(commissions.get(co_owner_2, 0) + net_value * 0.0095, 2)

# Convertir el diccionario a un DataFrame
commission_df = pd.DataFrame(commissions.items(), columns=["Sales Owner", "Commission (â‚¬)"])

# Ordenar por comisiÃ³n en orden descendente
commission_df = commission_df.sort_values(by="Commission (â‚¬)", ascending=False)

commission_df
```

### Test 5: DataFrame of Companies with Sales Owners

Provide a DataFrame (`df_3`) containing the following columns:

| Column           | Description                                                                                                                                                                                                                                       |
|------------------|---------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| company_id       | The `company_id` field must contain the unique identifier of the company.                                                                                                                                                                         |
| company_name     | The `company_name` field must contain the name of the company.                                                                                                                                                                                    |
| list_salesowners | The `list_salesowners` field should contain a unique and comma-separated list of salespeople who have participated in at least one order of the company. Please ensure that the list is sorted in ascending alphabetical order of the first name. |

Provide a DataFrame (`df_3`) containing the following columns:
- | company_id       | The `company_id` field must contain the unique identifier of the company
-  | company_name     | The `company_name` field must contain the name of the company
- | list_salesowners | The `list_salesowners` field should contain a unique and comma-separated list of salespeople who have participated in at least one order of the company. Please ensure that the list is sorted in ascending alphabetical order of the first name. |


```python
company_salesowners = merged_raw_df[['company_name', 'companyId', 'order_id', 'Main_Owner', 'Co_owner_1', 'Co_owner_2']].copy()
company_salesowners
```

```python
company_salesowners['company'] = company_salesowners['company_name'].str.lower()
company_salesowners['company'] = company_salesowners['company'].str.replace('.', '')

company_salesowners
```

```python
from rapidfuzz import process

# Unificar nombres de compaÃ±Ã­a con similitud de texto
def unify_company_names(df, threshold=80):
    unique_names = {}
    for name in df["company_name"].unique():
        match = process.extractOne(name, unique_names.keys(), score_cutoff=threshold)
        if match is not None:
            print(match)
            matched_name, score, _ = match
            unique_names[name] = matched_name
        else:
            unique_names[name] = name
    
    df["company"] = df["company_name"].map(unique_names)
    return df

company_salesowners = unify_company_names(company_salesowners)
company_salesowners
```

```python
# Crear diccionario para agrupar salesowners por empresa
company_dict = {}
for _, row in company_salesowners.iterrows():
    company_id = row["companyId"]
    company_name = row["company"]
    salesowners = set(filter(lambda x: x and x.lower() != "none", (', '.join(map(str, [row["Main_Owner"], row["Co_owner_1"], row["Co_owner_2"]]))).split(', ')))
    
    if company_id not in company_dict:
        company_dict[company_id] = {"company": company_name, "salesowners": salesowners}
    else:
        company_dict[company_id]["salesowners"].update(salesowners)

# Convertir diccionario a DataFrame
company_data = []
for company_id, data in company_dict.items():
    company_data.append({
        "id": company_id,
        "company": data["company"],
        "list_salesowners": ', '.join(sorted(data["salesowners"]))
    })

df_grouped = pd.DataFrame(company_data)

df_grouped
```
